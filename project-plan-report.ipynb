```python
# VG sales total by year
vg_sales_mod2 = vg_sales_mod.groupby('Year').sum().reset_index().drop('Rank', axis = 1) 
vg_sales_mod2 # Rank is the sum of all the ranks, can probably drop
vg_sales_mod3 = vg_sales_mod.groupby('Genre').sum().reset_index().sort_values(by = 'Global_Sales').drop(
    'Rank', axis = 1) #Sorted in Global Sales order
vg_sales_mod3 # Rank is the sum of all the ranks, can probably drop
vg_sales_mod4 = vg_sales_mod.groupby(['Year', 'Genre']).count().drop(columns = ['Name', 'Platform', 'Publisher', 'NA_Sales', 
                                                                                'EU_Sales', 'JP_Sales', 'Other_Sales', 'Global_Sales']).reset_index().pivot(index = 'Year',columns = 'Genre',values = 'Rank')
vg_sales_mod4 = vg_sales_mod4.fillna(0).reset_index()
# vg_sales_mod4
vg_sales_mod5 = vg_sales_mod.groupby(['Platform', 'Year']).sum().reset_index().drop('Rank', axis = 1)
vg_sales_mod5
vg_sales_mod6 = vg_sales_mod5.sort_values('Global_Sales', ascending = False)
vg_sales_mod6
```

```python
alt.data_transformers.disable_max_rows()
alt.Chart(vg_sales_mod2).mark_bar().encode(
    x=alt.X('Year'),
    y=alt.Y('Global_Sales') #sum sales
)
```
